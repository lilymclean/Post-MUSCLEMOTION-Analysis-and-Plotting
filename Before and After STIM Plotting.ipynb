```python
#Use this notebook to create plots of muscle contraction videos analyzed in ImageJ using MUSCLEMOTION

#This notebook requires the speed-of-contraction.txt files generated by MUSCLEMOTION

!pip install matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
onset_array = np.array([10000, 10000, 21000, 10000, 7000, 10000, 5000, 12000, 14000, 18000, 21000, 24000, 19000, 23000, 28000, 21000, 46000, 5000, 5000, 10000, 19000, 24000, 5000, 10000])

baseline = 50

def PlotContractionSpeed(title, txtfile, stim_onset):
    
  SpeedofContraction = np.loadtxt(txtfile) #Reads the .txt file and stores it in SpeedofContraction variable

  x = SpeedofContraction[:,0] #Indexes into the first column of the .txt file (time)
  y = SpeedofContraction[:,1] #Indexes into the second column of the .txt file (contraction speed)

  plt.plot(x, y) 

  plt.xlabel('Time (milliseconds)')

  plt.ylabel('Speed (au)')
    
  plt.title(title)

  plt.axhline(y=baseline, color='r', linestyle='-')

  plt.show()
    
  # Data difference
  ydiff = y[1:] - y[0:-1]

  # Find mean change
  ydiff_mean = np.abs(ydiff).mean()

  # Identify all indices greater than the mean
  spikes = ydiff > abs(ydiff_mean)+1

  where_spikes = np.where(spikes)[0]+1

  timestamps = x[where_spikes]

  timestamps = timestamps[timestamps<(2*stim_onset)]
    
  prestim_count = timestamps[timestamps<stim_onset]

  print(len(prestim_count))

  poststim_count = timestamps[timestamps>stim_onset]

  print(len(poststim_count))
```

```python
def PlotPreStim(title, txtfile, first_prestim_frame, last_prestim_frame):
    
  SpeedofContraction = np.loadtxt(txtfile) #Reads the .txt file and stores it in SpeedofContraction variable

  x = SpeedofContraction[first_prestim_frame:last_prestim_frame,0] 
  y = SpeedofContraction[first_prestim_frame:last_prestim_frame,1] 

  plt.plot(x, y) 

  plt.xlabel('Time (milliseconds)')

  plt.ylabel('Speed (au)')
    
  plt.axhline(y=baseline, color='r', linestyle='-')
    
  plt.title(title)

  plt.show()
    
  # Data difference
  ydiff = y[1:] - y[0:-1]

  # Find mean change
  ydiff_mean = np.abs(ydiff).mean()

  # Identify all indices greater than the mean
  spikes = ydiff > abs(ydiff_mean)+1

  where_spikes = np.where(spikes)[0]+1

  timestamps = x[where_spikes]
    
  amplitudes = y[where_spikes]
    
  peaks_above_baseline = amplitudes[np.where(amplitudes > baseline)] 
    
  num_peaks_above_baseline = (len(peaks_above_baseline))

  max_amplitude = np.max(amplitudes)
    
  average_amplitude = np.average(amplitudes)

  median_amplitude = np.median(amplitudes)

  print('Maximum amplitude =',max_amplitude)
    
  print ('Average amplitude =',average_amplitude)
 
  print('Median amplitude =',median_amplitude)

  print('Number of Peaks =', num_peaks_above_baseline)
```

```python
def PlotPostStim(title, txtfile, first_poststim_frame, last_poststim_frame):
    
  SpeedofContraction = np.loadtxt(txtfile) #Reads the .txt file and stores it in SpeedofContraction variable

  x = SpeedofContraction[first_poststim_frame:last_poststim_frame,0] 
  y = SpeedofContraction[first_poststim_frame:last_poststim_frame,1] 

  plt.plot(x, y) 

  plt.xlabel('Time (milliseconds)')

  plt.ylabel('Speed (au)')
    
  plt.title(title)

  plt.axhline(y=baseline, color='r', linestyle='-')

  plt.show()
    
  # Data difference
  ydiff = y[1:] - y[0:-1]

  # Find mean change
  ydiff_mean = np.abs(ydiff).mean()

  # Identify all indices greater than the mean
  spikes = ydiff > abs(ydiff_mean)+1

  where_spikes = np.where(spikes)[0]+1

  timestamps = x[where_spikes]
    
  amplitudes = y[where_spikes]
    
  peaks_above_baseline = amplitudes[np.where(amplitudes > baseline)] 
    
  num_peaks_above_baseline = (len(peaks_above_baseline))

  max_amplitude = np.max(amplitudes)
    
  average_amplitude = np.average(amplitudes)

  median_amplitude = np.median(amplitudes)

  print('Maximum Amplitude =',max_amplitude)
    
  print('Average Amplitude =',average_amplitude)

  print('Median Amplitude =',median_amplitude)

  print('Number of Peaks =', num_peaks_above_baseline)
```
