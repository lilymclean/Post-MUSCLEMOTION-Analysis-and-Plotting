```python
#Use this notebook to create plots of muscle contraction videos analyzed in ImageJ using MUSCLEMOTION

#This notebook requires the speed-of-contraction.txt files generated by MUSCLEMOTION

!pip install matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
#This cell defines a function (PlotContractionSpeed) to create plots of several videos at once. 

def PlotContractionSpeed(title, txtfile):
    
  SpeedofContraction = np.loadtxt(txtfile) #Reads the .txt file and stores it in SpeedofContraction variable

  x = SpeedofContraction[:,0] #Indexes into the first column of the .txt file (time)
  y = SpeedofContraction[:,1] #Indexes into the second column of the .txt file (contraction speed)

  plt.plot(x, y) 

  plt.xlabel('Time (milliseconds)')

  plt.ylabel('Speed (au)')
    
  plt.title(title)

  plt.show()
```

```python
#The code below follows the format of the PlotContractionSpeed function defined above.

#def PlotContractionSpeed(title, txtfile):

#Change the string variables below to reflect the title and .txt file you want to plot

#Examples shown below

PlotContractionSpeed('ChR2CS and POMC1 BL at 10 sec',
                     'ChR2CS and POMC1 BL at 10 sec.txt')

PlotContractionSpeed('ChR2CS and POMC1 DIV 17 BL 10sec-2',
                     'ChR2CS and POMC1 DIV 17 BL 10sec-2.txt')

PlotContractionSpeed('ChR2CS and POMC1 DIV 17 CTZ-1',
                     'ChR2CS and POMC1 DIV 17 CTZ-1.txt')

PlotContractionSpeed('ChR2CS and POMC1 NMJ DIV 12 spon',
                     'ChR2CS and POMC1 NMJ DIV 12 spon.txt')

PlotContractionSpeed('ChR2CS and POMC1 NMJ DIV 13 spontaneous BF',
                     'ChR2CS and POMC1 NMJ DIV 13 spontaneous BF.txt')
                     
PlotContractionSpeed('DIV 10 ChR2CS and POMC1 BL 10 sec.',
                     'DIV 10 ChR2CS and POMC1 BL 10 sec.txt')
                     
PlotContractionSpeed('DIV 10 ChR2CS and POMC1 BL 5 sec',
                     'DIV 10 ChR2CS and POMC1 BL 5 sec.txt')
```
