```python
#CTZ Stim
x = ['Before', 'After']
a = [0, 10]
b = [8, 10]
c = [0, 19]
d = [11, 22]
e = [17, 24]

avg_before_peaks = (a[0]+b[0]+c[0]+d[0]+e[0])/5
avg_after_peaks = (a[1]+b[1]+c[1]+d[1]+e[1])/5

avg_slope = avg_after_peaks-avg_before_peaks

print('Average Number of Contractions Pre-Stim',avg_before_peaks)
print('Average Number of Contractions Post-Stim',avg_after_peaks)
print('Average Slope',avg_slope)

avg = [avg_before_peaks, avg_after_peaks]

plt.plot(x, avg, color='red')
plt.plot(x, a, color='blue')
plt.plot(x, b, color='blue')
plt.plot(x, c, color='blue')
plt.plot(x, d, color='blue')
plt.plot(x, e, color='blue')

plt.ylabel('Number of Peaks')
    
plt.title('CTZ Stimulation')

plt.rcParams["figure.figsize"] = (2,5)

plt.show()


#LED Stim
x = ['Before', 'After']
a = [28, 22]
b = [6, 15]
c = [18, 13]
d = [5, 7]
e = [1, 9]
f = [1, 7]
g = [8, 17]
h = [9, 15]

avg_before_peaks = (a[0]+b[0]+c[0]+d[0]+e[0]+f[0]+g[0]+h[0])/8
avg_after_peaks = (a[1]+b[1]+c[1]+d[1]+e[1]+f[1]+g[1]+h[1])/8

avg_slope = avg_after_peaks-avg_before_peaks

print('Average Number of Contractions Pre-Stim',avg_before_peaks)
print('Average Number of Contractions Post-Stim',avg_after_peaks)
print('Average Slope',avg_slope)

avg = [avg_before_peaks, avg_after_peaks]

plt.plot(x, avg, color='red')
plt.plot(x, a, color='blue')
plt.plot(x, b, color='blue')
plt.plot(x, c, color='blue')
plt.plot(x, d, color='blue')
plt.plot(x, e, color='blue')
plt.plot(x, f, color='blue')
plt.plot(x, g, color='blue')
plt.plot(x, h, color='blue')

plt.ylabel('Number of Peaks')
    
plt.title('LED Stimulation')

plt.rcParams["figure.figsize"] = (2,5)

plt.show()


#Vehicle
x = ['Before', 'After']
a = [16, 14]
b = [20, 17]
c = [13, 6]
d = [32, 39]

avg_before_peaks = (a[0]+b[0]+c[0]+d[0])/4
avg_after_peaks = (a[1]+b[1]+c[1]+d[1])/4

avg_slope = avg_after_peaks-avg_before_peaks

print('Average Number of Contractions Pre-Stim',avg_before_peaks)
print('Average Number of Contractions Post-Stim',avg_after_peaks)
print('Average Slope',avg_slope)

avg = [avg_before_peaks, avg_after_peaks]

plt.plot(x, avg, color='red')
plt.plot(x, a, color='blue')
plt.plot(x, b, color='blue')
plt.plot(x, c, color='blue')
plt.plot(x, d, color='blue')

plt.ylabel('Number of Peaks')
    
plt.title('Vehicle')

plt.rcParams["figure.figsize"] = (2,5)

plt.show()


#No treatment (Spontaneous)
x = ['Before', 'After']
a = [19, 15]
b = [7, 8]
c = [9, 12]

avg_before_peaks = (a[0]+b[0]+c[0])/3
avg_after_peaks = (a[1]+b[1]+c[1])/3

avg_slope = avg_after_peaks-avg_before_peaks

print('Average Number of Contractions Pre-Stim',avg_before_peaks)
print('Average Number of Contractions Post-Stim',avg_after_peaks)
print('Average Slope',avg_slope)

avg = [avg_before_peaks, avg_after_peaks]

plt.plot(x, avg, color='red')
plt.plot(x, a, color='blue')
plt.plot(x, b, color='blue')
plt.plot(x, c, color='blue')

plt.ylabel('Number of Peaks')
    
plt.title('No Treatment')

plt.rcParams["figure.figsize"] = (2,5)

plt.show()


#Plain NMJ with CTZ Treatment
x = ['Before', 'After']
a = [17, 17]
b = [6, 4]
c = [23, 22]

avg_before_peaks = (a[0]+b[0]+c[0])/3
avg_after_peaks = (a[1]+b[1]+c[1])/3

avg_slope = avg_after_peaks-avg_before_peaks

print('Average Number of Contractions Pre-Stim',avg_before_peaks)
print('Average Number of Contractions Post-Stim',avg_after_peaks)
print('Average Slope',avg_slope)

avg = [avg_before_peaks, avg_after_peaks]

plt.plot(x, avg, color='red')
plt.plot(x, a, color='blue')
plt.plot(x, b, color='blue')
plt.plot(x, c, color='blue')

plt.ylabel('Number of Peaks')
    
plt.title('Plain NMJ with CTZ Stimulation')

plt.rcParams["figure.figsize"] = (2,5)

plt.show()
```
